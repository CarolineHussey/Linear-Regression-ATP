```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load and investigate the data here:
df=pd.read_csv(r'tennis_stats.csv')
print(df.head())
print(df.columns)
print(df.dtypes)
columns=df.columns.tolist()
print(columns)
```

```python
# Exploratory data analysis 
first=df[['FirstServe', 'FirstServePointsWon', 'FirstServeReturnPointsWon']]
first_columns = first.columns.tolist()
for c in first_columns:
    plt.scatter(first[c], df['Wins'])
    plt.title(c + " scatter")
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(c, fontsize=12)
    #print(c)
    plt.show()
    plt.clf()
second=df[['SecondServePointsWon', 'SecondServeReturnPointsWon']]
second_columns = second.columns.tolist()
for c in second_columns:
    plt.scatter(second[c], df['Wins'])
    plt.title(c + " scatter")
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(c, fontsize=12)
    #print(c)
    plt.show()
    plt.clf()
breaks=df[['BreakPointsConverted', 'BreakPointsFaced', 'BreakPointsOpportunities', 'BreakPointsSaved']]
breaks_columns = breaks.columns.tolist()
for c in breaks_columns:
    plt.scatter(breaks[c], df['Wins'])
    plt.title(c + " scatter")
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(c, fontsize=12)
    #print(c)
    plt.show()
    plt.clf()
```

```python
## single feature linear regressions
features = df[['BreakPointsOpportunities']]
outcome = df[['Winnings']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
model = LinearRegression()
model.fit(features_train,outcome_train)
sc=model.score(features_test,outcome_test)
print(sc)
prediction = model.predict(features_test)
plt.scatter(outcome_test,prediction, alpha=0.4)
plt.title('Prediction of Wins based on Break Point Opportunities')
plt.show()
plt.clf()
```

```python
## Two feature linear regression
features = df[['BreakPointsOpportunities', 'Aces']]
outcome = df[['Winnings']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
model = LinearRegression()
model.fit(features_train,outcome_train)
sc=model.score(features_test,outcome_test)
print(sc)
prediction = model.predict(features_test)
plt.scatter(outcome_test,prediction, alpha=0.4)
plt.title('Prediction of Wins based on Break Point Opportunities and Aces')
plt.show()
plt.clf()
```

```python
## multiple feature linear regression
features = df[['BreakPointsConverted','BreakPointsFaced','BreakPointsOpportunities']]
outcome = df[['Winnings']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
model = LinearRegression()
model.fit(features_train,outcome_train)
sc=model.score(features_test,outcome_test)
print(sc)
prediction = model.predict(features_test)
plt.scatter(outcome_test,prediction, alpha=0.4)
plt.title('Prediction of Wins based on Break Points Converted, Break Points Faced and Break Point Opportunities')
plt.show()
plt.clf()
```
